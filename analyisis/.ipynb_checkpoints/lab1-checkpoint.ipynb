### Funciones auxiliares

```python
import locale
import re

# Set the locale to Spanish
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

def spanish_month_name_to_number(month_name):
    date_object = datetime.strptime(month_name.lower(), "%B")
    return date_object.month

def paddingNumber(value):
    value = str(value)
    if len(value) == 1:
        return f"0{value}"
    return value

def convertToDate(value):
    try:
        date = pd.to_datetime(value)
        return date
    except ValueError:
        #print(f"Cannot convert {value} to Date")
        #Trying to transform invalid dates, we put None value to exclude in next steps
        return None

def getDayWeek(date):
    return date.isoweekday()

def getWeekDayType(day):
    # 0 = weekend
    # 1 = labor day
    weekend = [6,7]
    if day in weekend:
        return 1
    return 0

def getSeason(date):
    #https://es.scribd.com/document/359340681/Estaciones-Del-Ano-de-Guatem-Ln
    #0 = spring
    #1 = summer
    #2 = autumn
    #3 = winter 
    
    #spring
    start_date = datetime(2023, 3, 21)
    end_date = datetime(2023, 6, 20)
    if(start_date <= date <= end_date):
        return 0
    
    #summer
    start_date = datetime(2023, 6, 21)
    end_date = datetime(2023, 9, 23)
    if(start_date <= date <= end_date):
        return 1

    #autumn
    start_date = datetime(2023, 9, 24)
    end_date = datetime(2023, 12, 21)
    if(start_date <= date <= end_date):
        return 2
    
    #winter
    start_date = datetime(2023, 12, 22)
    end_date = datetime(2023, 12, 31)
    if(start_date <= date <= end_date):
        return 3
    
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 3, 20)
    if(start_date <= date <= end_date):
        return 3
    
def getSeason_dumy(date):
    #https://es.scribd.com/document/359340681/Estaciones-Del-Ano-de-Guatem-Ln
    #0 = spring
    #1 = summer
    #2 = autumn
    #3 = winter 
    
    #spring
    start_date = datetime(2022, 3, 21)
    end_date = datetime(2022, 6, 20)
    if(start_date <= date <= end_date):
        return 0
    
    #summer
    start_date = datetime(2022, 6, 21)
    end_date = datetime(2022, 9, 23)
    if(start_date <= date <= end_date):
        return 1

    #autumn
    start_date = datetime(2022, 9, 24)
    end_date = datetime(2022, 12, 21)
    if(start_date <= date <= end_date):
        return 2
    
    #winter
    start_date = datetime(2022, 12, 22)
    end_date = datetime(2022, 12, 31)
    if(start_date <= date <= end_date):
        return 3
    
    start_date = datetime(2022, 1, 1)
    end_date = datetime(2022, 3, 20)
    if(start_date <= date <= end_date):
        return 3

```

### Función que devuelve la data de validación fuera del rango del modelo

```python
def createData():
    datetime_range = pd.date_range(start='2022-01-01 00:00', end='2022-12-31 23:00', freq='H')
    df = pd.DataFrame(datetime_range, columns=['Datetime'])

    df['Date'] = df['Datetime'].dt.date
    df['Hora'] = df['Datetime'].dt.hour
    df['mes'] = df['Datetime'].dt.month
    df['dia'] = df['Datetime'].dt.day
    df['object_date'] =  df['Date'].apply(convertToDate)
    
    
    df['num_day'] = df['Date'].apply(getDayWeek)
    df['IsWeekend'] = df['num_day'].apply(getWeekDayType)   

    df['validHour'] = (df['Hora'] >= 7) &  (df['Hora'] <= 19)
    df['validHour'] = df['validHour'].astype(int)
    
    holidays = [date(2022, 1, 1), date(2022, 4, 11),date(2022, 4, 12),date(2022, 4, 13),date(2022, 4, 14),date(2022, 4, 15),date(2022, 4, 16),
                date(2022, 4, 17),date(2022, 5, 1),date(2022, 7, 4),date(2022, 8, 15),date(2022, 9, 15),date(2022, 10, 20),date(2022, 11, 1),
                date(2022, 12, 31), date(2022, 12, 25),date(2022, 12, 24)]
    holiday_set = set(holidays)

    start_date = '2022-01-14'
    end_date = '2022-11-30'
    date_range = pd.date_range(start=start_date, end=end_date)
    df['private_school'] = df['Date'].isin(date_range)
    df['private_school'] = df['private_school'].astype(int)

    start_date = '2022-02-21'
    end_date = '2022-11-30'
    date_range = pd.date_range(start=start_date, end=end_date)
    df['public_school'] = df['object_date'].isin(date_range)
    df['public_school'] = df['public_school'].astype(int)

    #seasons
    df['season'] =  df['object_date'].apply(getSeason_dumy)

    df['IsHoliday'] = df['object_date'].isin(holiday_set)
    df['IsHoliday'] = df['IsHoliday'].astype(int)


    df['Year'] = df['Datetime'].dt.year
    df['Month'] = df['Datetime'].dt.month
    df['Day'] = df['Datetime'].dt.day

    # Group by year and month, and then find the day of the month for each day
    df['Day_of_month'] = df.groupby(['Year', 'Month'])['Day'].transform('min')
    df['Week_month'] = ((df['Day'] - df['Day_of_month']) // 7) + 1
    
    if MODE == 'cyclical':
        df['day_week_sin'] = np.sin(2 * np.pi * df['num_day'] / 7)
        df['day_week_cos'] = np.cos(2 * np.pi * df['num_day'] / 7)
        df['Hora'] = df['Hora'].astype(int)
        df['hour_sin'] = np.sin(2 * np.pi * df['Hora'] / 23)
        df['hour_cos'] = np.cos(2 * np.pi * df['Hora'] / 23)
        df['Week_month_sin'] = np.sin(2 * np.pi * df['Week_month'] / 5)
        df['Week_month_cos'] = np.cos(2 * np.pi * df['Week_month'] / 5)
        df['mes_sin'] = np.sin(2 * np.pi * df['mes'] / 12)
        df['mes_cos'] = np.cos(2 * np.pi * df['mes'] / 12)      
    
    df = df.drop(['Datetime',
                                    'object_date',
                                    'Day_of_month',
                                    'Year',
                                    'Month',
                                    'Day'
    ], axis=1)
    #print(df)
    df.to_excel("testing_dummy.xlsx")   
    return df
```

```python
def formatMessage(sorted_list):
    middle_index = len(sorted_list) // 2    
    first_half = sorted_list[:middle_index]
    second_half = sorted_list[middle_index:]
    
    print()
    print("Elementos con mayor error")
    for element in first_half:
        output = []
        for key, value in element.items():
            output.append(f"{key}: {value}")   
        print(", ".join(output))
    
    print()
    print("Elementos con menor error")
    for element in second_half:
        output = []
        for key, value in element.items():
            output.append(f"{key}: {value}")   
        print(", ".join(output))
```

### Función que realiza la lectura completa del archivo de entrada hoja por hoja

```python
def readFileComplete():
    dfs = pd.read_excel(excel_file, sheet_name=None,  skiprows=2)
    df = pd.read_excel(excel_file, header=start_row - 1)
    df.columns = df.iloc[0]

    # Display the DataFrame from each sheet
    df_complete = pd.DataFrame()
    for sheet_name, df in dfs.items():
        print(f"Data from sheet '{sheet_name}':")
        final_df = df[0:end_row]
        final_df = final_df.drop(df.columns[1], axis=1)
        final_df = final_df.drop(['Hora'], axis=1)
        df_transpuesto2 = final_df.transpose()
        df_transpuesto2.reset_index(inplace=True)
        melted_df = df_transpuesto2.melt(id_vars=['index'], value_name='Energy')
        melted_df["mes"] = sheet_name
        melted_df['mes'] = melted_df['mes'].apply(spanish_month_name_to_number)
        df_transpuesto = melted_df
        df_complete = pd.concat([df_complete,df_transpuesto])
    return df_complete
```

### Funcion que permite realizar todas las transformaciones necesarias que seran inyectadas en el modelo
Transformaciones aplicadas

- Dia de la semana, valor en un rango de 1 a 7.
- Es fin de semana, valor binario que determina si el día es sábado o domingo.
- Horas validas, valor en un rango de 7 a 19 que fueron definidos como horario valido de trabajo.
- Escuela pública, valor binario que determina si está en un periodo escolar del sector público.
- Escuela privada, valor binario que determina si está en un periodo escolar del sector privado.
- Es feriado, valor binario que determina si la fecha es una asueto o feriado.
- Temporada, valor en un rango de 1 a 4 que determina si es: verano, invierno, otoño o primavera.
- Energía, debido al análisis de distribución se aplicó la raíz cuadrada.
- Variables temporales, se aplicó codificación cíclicas para las variables que son periódicas [día de la semana, mes, semana del mes, hora]


```python
def transformDataSet(df_complete):
    remove_specific_columns = []
    df_complete = df_complete.rename(columns={'index': 'dia', 'variable':'Hora'})
    df_complete['dia_str'] = df_complete['dia'].apply(paddingNumber)
    df_complete['mes_str'] = df_complete['mes'].apply(paddingNumber)
    df_complete['Date'] =f"{str(year)}-" + df_complete['mes_str'].astype(str) +'-' + df_complete['dia_str'].astype(str)
    df_complete['object_date'] =  df_complete['Date'].apply(convertToDate)
    df_complete.reset_index(drop=True, inplace=True)

    valid_mask = df_complete['object_date'].notna()
    df_complete = df_complete[valid_mask]
    df_complete['num_day'] = df_complete['object_date'].apply(getDayWeek)
    df_complete['IsWeekend'] = df_complete['num_day'].apply(getWeekDayType)   


    df_complete['validHour'] = (df_complete['Hora'] >= 7) &  (df_complete['Hora'] <= 19)
    df_complete['validHour'] = df_complete['validHour'].astype(int)

    holidays = [date(2023, 1, 1), date(2023, 4, 2),date(2023, 4, 3),date(2023, 4, 4),date(2023, 4, 5),date(2023, 4, 6),date(2023, 4, 7),
                date(2023, 4, 8),date(2023, 4, 9),date(2023, 7, 3),date(2023, 8, 15),date(2023, 9, 15),date(2023, 10, 20),date(2023, 11, 1),
                date(2023, 12, 31), date(2023, 12, 25),date(2023, 12, 24)]
    holiday_set = set(holidays)

    start_date = '2023-01-06'
    end_date = '2023-10-30'
    date_range = pd.date_range(start=start_date, end=end_date)
    df_complete['private_school'] = df_complete['object_date'].isin(date_range)
    df_complete['private_school'] = df_complete['private_school'].astype(int)

    start_date = '2023-02-15'
    end_date = '2023-11-30'
    date_range = pd.date_range(start=start_date, end=end_date)
    df_complete['public_school'] = df_complete['object_date'].isin(date_range)
    df_complete['public_school'] = df_complete['public_school'].astype(int)

    #seasons
    df_complete['season'] =  df_complete['object_date'].apply(getSeason)


    df_complete['IsHoliday'] = df_complete['object_date'].isin(holiday_set)
    df_complete['IsHoliday'] = df_complete['IsHoliday'].astype(int)


    df_complete['Year'] = df_complete['object_date'].dt.year
    df_complete['Month'] = df_complete['object_date'].dt.month
    df_complete['Day'] = df_complete['object_date'].dt.day

    # Group by year and month, and then find the day of the month for each day
    df_complete['Day_of_month'] = df_complete.groupby(['Year', 'Month'])['Day'].transform('min')
    df_complete['Week_month'] = ((df_complete['Day'] - df_complete['Day_of_month']) // 7) + 1

    df_complete = df_complete.drop(['dia_str',
                                    'mes_str',
                                    'object_date',
                                    'Day_of_month',
                                    'Year',
                                    'Month',
                                    'Day'
    ], axis=1)
    
    if APPLY_TRANSFORMS:
        df_complete['Energy'] = np.sqrt(df_complete['Energy'])
    
    if MODE == 'cyclical':
        df_complete['day_week_sin'] = np.sin(2 * np.pi * df_complete['num_day'] / 7)
        df_complete['day_week_cos'] = np.cos(2 * np.pi * df_complete['num_day'] / 7)
        df_complete['Hora'] = df_complete['Hora'].astype(int)
        df_complete['hour_sin'] = np.sin(2 * np.pi * df_complete['Hora'] / 23)
        df_complete['hour_cos'] = np.cos(2 * np.pi * df_complete['Hora'] / 23)
        df_complete['Week_month_sin'] = np.sin(2 * np.pi * df_complete['Week_month'] / 5)
        df_complete['Week_month_cos'] = np.cos(2 * np.pi * df_complete['Week_month'] / 5)
        df_complete['mes_sin'] = np.sin(2 * np.pi * df_complete['mes'] / 12)
        df_complete['mes_cos'] = np.cos(2 * np.pi * df_complete['mes'] / 12)
        
        remove_specific_columns = remove_specific_columns + ['Hora','num_day','Week_month', 'mes']
    else:
        remove_specific_columns = []
        
    df_complete.to_excel("output.xlsx")      
    return df_complete, remove_specific_columns 
    
```

### Función que grafica la relación entre la variable predictora y la variable independiente

```python
def showScatterPlotBetweenXandY(dataset,dependant):
    x_columns = dataset.columns 
    y_column = dependant 

    plt.figure(figsize=(12, 8))
    index_col = 1
    for i, x in enumerate(x_columns,1):
        if(x == dependant):
            continue
        correlation_coefficient = round(dataset[x].corr(dataset[y_column], method='pearson'),4)
        plt.subplot(4, 4, index_col)
        sns.scatterplot(data=dataset, x=x, y=y_column)
        plt.title(f'CR={correlation_coefficient}')
        index_col=index_col+1
    plt.tight_layout()
    plt.show()
```

### Función que permite generar un modelo polinomial

```python
def polinomialEvaluate(x_train, y_train, x_test, print_result=True):
    poly_features = PolynomialFeatures(degree=2)
    x_train_p = poly_features.fit_transform(x_train)
    x_test_p = poly_features.transform(x_test)
    model = LinearRegression() 
    model.fit(x_train_p,y_train)     
    if print_result:
        coeff = {key: value for key, value in zip(poly_features.get_feature_names_out(), model.coef_[0])}
        print(coeff)
    return model, x_test_p
```

### Función que permite generar un modelo utilizando regularización

```python
def pilelineRidge(x_train, y_train, x_test, print_result=True):
    pipiline_ridge = Pipeline(
    [
     ('scaler', StandardScaler()),
     ('poly', PolynomialFeatures(degree=2)),
     ('ridge',Ridge(alpha=10, max_iter=1000))    
    ])
    
    pipiline_ridge.fit(x_train, y_train)
    if print_result:
        coeff = {key: value for key, value in zip(pipiline_ridge[:-1].get_feature_names_out(), pipiline_ridge['ridge'].coef_[0])}
        print(coeff)
    return pipiline_ridge
    
```

### Función que permite generar los valores de correlación entra las variables

```python
def getCorrelationMatrix(df_complete):
    corr_matrix = df_complete.corr(method='pearson')
    tril = np.tril(np.ones(corr_matrix.shape)).astype(bool)
    corr_matrix[tril] = np.nan
    corr_matrix_tidy = corr_matrix.stack().reset_index(name='r')
    corr_matrix_tidy['r'] = corr_matrix_tidy['r'].round(4)
    corr_matrix_tidy = corr_matrix_tidy.rename(columns={'level_0': 'variable_1', 'level_1': 'variable_2'})
    corr_matrix_tidy = corr_matrix_tidy.dropna()
    corr_matrix_tidy['r_abs'] = corr_matrix_tidy['r'].abs()
    corr_matrix_tidy = corr_matrix_tidy.sort_values('r_abs', ascending=False).reset_index(drop=True)
    return corr_matrix_tidy, corr_matrix
```

### Función que permite generar los valores de correlación entra las variables, únicamente mostrando un lado de la grafica

```python
def correlationCoefficientesHalfView(corr_matrix):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
    sns.heatmap(
        corr_matrix,
        annot     = True,
        cbar      = False,
        annot_kws = {"size": 8},
        vmin      = -1,
        vmax      = 1,
        center    = 0,
        cmap      = "viridis",
        square    = True,
        ax        = ax
    )
    ax.set_xticklabels(
        ax.get_xticklabels(),
        rotation = 45,
        horizontalalignment = 'right',
    )
    ax.tick_params(labelsize = 10)
```

### Función que permite visualizar los datos de distribución de forma resumida.

```python
def graphDistribution(df_complete):
    fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(9, 6))
    axes = axes.flat
    columnas_numeric = df_complete.select_dtypes(include=np.number).columns
    print(len(columnas_numeric))

    for i, colum in enumerate(columnas_numeric):
        sns.histplot(
            data    = df_complete,
            x       = colum,
            stat    = "count",
            kde     = True,
            color   = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
            ax      = axes[i]
        )
        axes[i].set_title(colum, fontsize = 10)
        axes[i].tick_params(labelsize = 8)
        axes[i].set_xlabel("")

    fig.tight_layout()
    plt.subplots_adjust(top = 0.9)
    fig.suptitle('Distribución variables numéricas', fontsize = 12)
```

### Función que permite calcular el gradiente del error y graficar la barra para ver la comparativa entre la data real y la de predicción

```python
def testingModels(y_test,y_predict, graph = True):
    error = np.sum(y_predict - y_test)**2
    error = round((error/len(y_test)) ** 0.5, 2)
    
    if graph:
        y_test_value = round(np.sum(y_test),2)
        y_predict_value = round(np.sum(y_predict),2)
        plt.figure(figsize=(8, 6))
        plt.bar(['test','predict'], [y_test_value,y_predict_value],color=['blue', 'green'])
        plt.xlabel('Source')
        plt.ylabel('Values')
        plt.title(f"Test: {y_test_value} vs Predict: {y_predict_value}, Error: {error} ")
        plt.grid(True)
        plt.show()
    return error
```

### Función que permite evaluar el modelo calculando las variables que permitan verificar el modelo.

```python
def evaluateModel(model, x, y, cv=5):
    y_predict = model.predict(x)
    mae = mean_absolute_error(y, y_predict)
    mse = mean_squared_error(y, y_predict)
    rmse = np.sqrt(mse)
    scores = cross_val_score(model, x,y, cv=cv, scoring='neg_mean_absolute_error')
    return {
        'MAE': round(mae,2),
        'MSE': round(mse,2),
        'RMSE': round(rmse,2),
        'CV MAE': round(-np.mean(scores),2),
        'R2 Score': round(r2_score(y, y_predict),2)
    }

                             
```

### Función que permite generar los modelos y comparar
- Se genera el modelo lineal
- Se genera el modelo polinomial
- Se genera el modelo polinomial regularizado


```python
def compareModels(col_training):
    col_training = col_training[::-1]
    print(col_training)
    col_size=len(col_training)

    for i in range(0,len(col_training)):
        removeItems = ['Energy','Date'] + col_training[:(col_size - (col_size - i))]
        print("Removing columns of train: " + ", ".join(removeItems))
        X = df_complete.drop(columns=['Energy','Date'] + remove_specific_columns + col_training[:(col_size - (col_size - i))])
        y = df_complete['Energy']

        X_train, X_test, y_train, y_test = train_test_split(
                                                X,
                                                y.values.reshape(-1,1),
                                                train_size   = 0.8,
                                                random_state = 1234,
                                                shuffle      = True
                                            )
        print(f"{'*' * 50} Linear {'*'*50}")
        model = LinearRegression() 
        model.fit(X_train,y_train)
        results = evaluateModel(model, X_test, y_test)
        coeff = {key: value for key, value in zip(X.columns.tolist(), model.coef_[0])}
        print(coeff)
        print(results)

        print(f"{'*' * 50} Polynomial {'*'*50}")
        model, x_test_p = polinomialEvaluate(X_train,y_train, X_test)
        results = evaluateModel(model, x_test_p, y_test)
        print(results)
        
        print(f"{'*' * 50} Ridge {'*'*50}")
        pipiline_ridge = pilelineRidge(X_train,y_train, X_test)
        results = evaluateModel(pipiline_ridge, X_test, y_test)
        print(results)   

        print()       
```

```python
TOP = 3
### Verificando el modelo por horas
print("Analizando el modelo por horas")
horas = (np.arange(0, 24)).tolist()
results_list = []
for i in range(len(horas)):
    hora_sen = np.sin(2 * np.pi * horas[i] / 23)
    hora_cos = np.cos(2 * np.pi * horas[i] / 23)
    filterAux = df_complete[(df_complete['hour_sin'] == hora_sen) & (df_complete['hour_cos'] == hora_cos)]
    X = filterAux.drop(columns=['Energy','Date'] + remove_specific_columns)
    y = filterAux['Energy']   
    y = y.values.reshape(-1,1)

    y_predict = pipiline_ridge.predict(X)
    error = testingModels(y,y_predict, False)
    results_list.append({'hora': horas[i],'error': error})

sorted_list = sorted(results_list, key=lambda x: x['error'], reverse=True)
formatMessage(sorted_list)
print()


### Verificando el modelo opr horas y mes
print("Analizando el modelo por horas y mes")
meses =  (np.arange(1, 13)).tolist()
best_by_hour = []
results_list = []
for i in range(len(horas)):
    error_aux = 0
    for j in range(len(meses)):
        mes_sin = np.sin(2 * np.pi * meses[j] / 12)
        mes_cos = np.cos(2 * np.pi * meses[j] / 12)
        hora_sen = np.sin(2 * np.pi * horas[i] / 23)
        hora_cos = np.cos(2 * np.pi * horas[i] / 23)

        filterAux = df_complete[(df_complete['hour_sin'] == hora_sen) & (df_complete['hour_cos'] == hora_cos) 
                               & (df_complete['mes_sin'] == mes_sin) ^ (df_complete['mes_cos'] == mes_cos)]
        X = filterAux.drop(columns=['Energy','Date'] + remove_specific_columns)
        y = filterAux['Energy']   
        y = y.values.reshape(-1,1)

        y_predict = pipiline_ridge.predict(X)
        error = testingModels(y,y_predict, False)
        error_aux = error_aux + error
    results_list.append({'hora': horas[i], 'error': round(error_aux/len(meses),2)})
sorted_list = sorted(results_list, key=lambda x: x['error'], reverse=True)
formatMessage(sorted_list)
```
