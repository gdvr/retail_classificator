```python
def read_text_file(file_path,rows=0):
    rowNumber = extractWeekNumber(file_path)
    data = pd.read_excel(file_path,skiprows=[i for i in range(0,rows-1)])
    if (rowNumber):
        data['Semana'] = rowNumber
    #print(data.head)
    return data
```

```python
def extractWeekNumber(path):
    match = re.search(r"semana_(\d+)", path)
    if match:
        week_number = match.group(1)
        return week_number
    else:
        print("No week number found in the file path.")
```

```python
def main(path):
    print(path)
    data = []
    os.chdir(path)
    files = ""
    for file in os.listdir():
        if file.endswith(".xlsx"):
            file_path = f"{path}\{file}"
            if "output" in file_path.lower():
                print(f"Processing {file_path}...")
                data.append(read_text_file(file_path))
                print(f"Processed {file_path}")
    if len(data) != 0:
        data = pd.concat(data)
    else:
        print('Error, debe tener el archivo datos en xlsx y archivo ecat en txt')

    return data
```

```python
df_test = data.copy()

# Check for duplicate indices
if not df_test.index.is_unique:
    print("Duplicate indices found.")
    print(df_test.index[df_test.index.duplicated()])

# Reset the index to ensure uniqueness (optional)
df_test = df_test.reset_index(drop=True)
```

```python
# Sort the data by 'storeId', 'productId', and 'Fecha'
df_test = df_test.sort_values(by=['storeId', 'productId', 'Fecha']).reset_index(drop=True)

# Define a function to calculate the rolling average for each group
def calculate_last_4_weeks_avg(group):
    # Shift the column to exclude the current row in the rolling calculation
    shifted_venta = group["Venta dia anterior"].shift(1)
    
    # Calculate the rolling average on the shifted column
    last_4_weeks_avg = (
        shifted_venta.rolling(window=4, min_periods=1).mean()
    )
    # Ensure at least 4 valid rows for rolling average, otherwise set to -1
    group["Last 4 Weeks Avg"] = last_4_weeks_avg.where(
        shifted_venta.rolling(window=4, min_periods=4).count() >= 4, -1
    )
    group['Last 4 Weeks Avg'].fillna(-1, inplace=True)
    return group

# Apply the rolling average calculation group-wise
df_test = df_test.groupby(['storeId', 'productId'], group_keys=False).apply(calculate_last_4_weeks_avg)

# Display results
print(df_test[['storeId', 'productId', 'Fecha', 'Venta dia anterior', 'Last 4 Weeks Avg']])
```

```python
df_test = df_test.sort_values(by='Fecha').reset_index(drop=True)

# Create a 'Last Sale Date' using the condition: Venta dia anterior > 0 and Fecha < current row's Fecha
df_test['Last Sale Date'] = df_test.apply(
    lambda row: df_test.loc[(df_test['Venta dia anterior'] > 0) & (df_test['Fecha'] < row['Fecha']), 'Fecha'].max() 
    if df_test.loc[(df_test['Venta dia anterior'] > 0) & (df_test['Fecha'] < row['Fecha']), 'Fecha'].size > 0 
    else pd.NaT, axis=1
)

# Calculate the 'Days Since Last Sale'
df_test['Days Since Last Sale'] = (df_test['Fecha'] - df_test['Last Sale Date']).dt.days

# Handle NaT (for the first row or rows with no previous sales) by filling with a placeholder value (e.g., -1)
df_test['Days Since Last Sale'] = df_test['Days Since Last Sale'].fillna(-1)

# Display the result
print(df_test[['Fecha', 'Last Sale Date', 'Days Since Last Sale']])
```

```python
# Sort the data by 'Fecha', 'storeId', and 'productId' for proper grouping
df_test = df_test.sort_values(by=['storeId', 'productId', 'Fecha']).reset_index(drop=True)

# Define a function to calculate 'Last Sale Date' and 'Days Since Last Sale' within each group
def calculate_last_sale_date(group):
    # Create a shifted 'Fecha' for rows where 'Venta dia anterior' > 0
    group['Last Sale Date'] = group.loc[group['Venta dia anterior'] > 0, 'Fecha'].shift(1)
    # Forward-fill 'Last Sale Date' within the group
    group['Last Sale Date'] = group['Last Sale Date'].ffill()
    # Calculate 'Days Since Last Sale'
    group['Days Since Last Sale'] = (group['Fecha'] - group['Last Sale Date']).dt.days
    return group

# Apply the function group-wise by 'storeId' and 'productId'
df_test = df_test.groupby(['storeId', 'productId'], group_keys=False).apply(calculate_last_sale_date)

# Display relevant columns for verification
print(df_test[['storeId', 'productId', 'Fecha', 'Venta dia anterior', 'Last Sale Date', 'Days Since Last Sale']])
```

```python
# Filter rows where 'Stock dia actual' > 0
df_with_stock = df_test[df_test['Stock dia actual'] > 0]

# Group by 'productId', 'País', and 'Fecha', counting unique 'storeId'
presence_counts = (
    df_with_stock.groupby(['productId', 'País', 'Fecha'])['storeId']
    .nunique()  # Count unique stores
    .reset_index(name='Total Presence')  # Rename column to 'Total Presence'
)

# Merge the "Total Presence" back into the original DataFrame
df_test = df_test.merge(
    presence_counts, 
    on=['productId', 'País', 'Fecha'], 
    how='left'
)

# Fill NaN values in 'Total Presence' with 0 (for rows without stock > 0 on a given date)
df_test['Total Presence'] = df_test['Total Presence'].fillna(0).astype(int)

# Display relevant columns for verification
print(df_test[['productId', 'País', 'Fecha', 'storeId', 'Stock dia actual', 'Total Presence']])
```

```python
# Define a function to calculate rotation type
def calculate_rotation_type(group):
    # Calculate the mode of 'Days Since Last Sale'
    mode_value = group['Days Since Last Sale'].mode()
    
    # If mode exists, classify based on its value
    if not mode_value.empty:
        mode_value = mode_value[0]  # Take the first mode if multiple modes exist
        if 1 <= mode_value <= 15:
            group['Rotation Type'] = 'A'
        elif 16 <= mode_value <= 60:
            group['Rotation Type'] = 'B'
        elif mode_value > 90:
            group['Rotation Type'] = 'C'
        else:
            group['Rotation Type'] = 'Unknown'
    else:
        # If no mode can be calculated, assign 'Unknown'
        group['Rotation Type'] = 'Unknown'
    
    return group

# Apply the calculation group-wise
df_test = df_test.groupby(['storeId', 'productId'], group_keys=False).apply(calculate_rotation_type)

# Display results
print(df_test[['storeId', 'productId', 'Days Since Last Sale', 'Rotation Type']])
```

```python
def calculate_percentage_diff_and_condition(group):
    # Calculate the average of the last 4 weeks excluding the current week
    group['Last 4 Weeks Avg (Excl Current)'] = (
        group['Venta dia anterior']
        .shift(1)  # Shift to exclude the current week
        .rolling(window=4, min_periods=4)  # Consider only full 4 weeks
        .mean()
    )
    
    # Calculate the percentage difference
    group['Percentage Difference'] = (
        (group['Venta dia anterior'] / group['Last 4 Weeks Avg (Excl Current)']) - 1
    )
    
    # Determine thresholds based on 'Rotation Type'
    thresholds = {
        'A': 0.25,
        'B': 0.40,
        'C': 0.50
    }
    group['Condition'] = group.apply(
        lambda row: row['Percentage Difference'] > thresholds.get(row['Rotation Type'], float('inf')),
        axis=1
    )
    
    return group

# Apply the logic group-wise by 'storeId' and 'productId'
df_test = df_test.groupby(['storeId', 'productId'], group_keys=False).apply(calculate_percentage_diff_and_condition)

# Display the results
print(df_test[['storeId', 'productId', 'Fecha', 'Venta dia anterior', 'Last 4 Weeks Avg (Excl Current)', 'Percentage Difference', 'Rotation Type', 'Condition']])
```
